```python
import pandas as pd

fruit_data = {
    'apple_sales': [3, 2, 0, 2, 9, 4, 7, 2],
    'orange_sales': [0, 3, 2, 7, 7, 8, 3, 0]
}

# Fix 1: RangeIndex only needs start and stop (and optionally step)
# This creates an index from 2000 to 2007
idx = pd.RangeIndex(start=2000, stop=2008)

# Fix 2: Use the 'idx' variable you just created for the DataFrame index
purchases = pd.DataFrame(fruit_data, index=idx)

print(purchases)
```

```python
rows,cols=purchases.shape
print(f'Rows:{rows}')
print(f'cals:{rows}')
```
